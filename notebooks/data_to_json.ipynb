```python
df = pd.read_csv("data/base/item_lists.csv", sep=";")
n_cols = len(df.columns)
df.head()
```

```python
item_lists = {}

for k, i in enumerate(range(0, int(n_cols), 2)):
    item_list = df.iloc[:, i:i+2].dropna()
    item_list =  item_list.drop(0, axis=0).values
    
    item_lists[k] = {}
    for item, qt in item_list:
        item_lists[k][item] = int(qt)
    print(item_lists)

with open("data/item_lists.json", 'w') as outfile:
    json.dump(item_lists, outfile)
```

```python
df = pd.read_csv("data/base/item_locations.csv", sep=";")
df.head()
```

```python
items = {}

for item, location in df.groupby("item", as_index=False).agg(list).values:
    items[item] = location

with open("data/items.json", 'w') as outfile:
    json.dump(items, outfile)
```
